```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
%matplotlib inline
```

```python
dt = pd.read_csv('./citrus.csv')
dt.head(5)
```

# Description
### This dataset takes the color, weight, and diameter of an "average" orange and generates a larger dataset containing a wide variety of values.



```python
dt.info()
```

```python
dt.shape
```

```python
dt.describe()
```

```python

orange_dt=dt.sample(30) #because i have alot of data rows so i choose 30 random from it
orange_dt
X = orange_dt['diameter'].values
y = orange_dt['weight'].values
```

```python
X=X.reshape(-1,1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X , y , test_size = 0.25, random_state = 4 )
```

```python
model = LinearRegression()
```

```python

model.fit(X_train, y_train)
```

# Regression Modelâ€™s Score

```python
accu = model.score(X_test , y_test)
accu*100
```

```python

y_predict = model.predict(X_test)
```

# Conclusions 

```python
plt.scatter(X_train, y_train)
plt.plot(X_test, y_predict , color = 'orange')
plt.xlabel('Diameter (cm)')
plt.ylabel('Weight (g)')
plt.show()
```

```python
plt.scatter(X_test, y_test)
plt.plot(X_test, y_predict , color = 'orange')
plt.xlabel('Diameter (cm)')
plt.ylabel('Weight (g)')
plt.show()
```

```python
orange_diameter = 7
orange_weight = model.predict([[orange_diameter]])[0]
print(f'If the diameter of orange = {orange_diameter} cm. The weight for the orange = {orange_weight} g. ')
```
